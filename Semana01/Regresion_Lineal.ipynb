# Regresión lineal

Este documento es para mostrar un ejemplo de cómo se realiza una regresión lineal en Python.

Para este ejemplo, se utilizará el set de datos [Linear Regression]("https://www.kaggle.com/datasets/andonians/random-linear-regression").

## Importación de datos

```python
# Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
#Lectura de datos
data = pd.read_csv('./test.csv')
```

```python
data[['x', 'y']].head()
```

```python
data.plot.scatter(x='x',y='y')
plt.show()
```

## Cálculo de los datos

Calculamos la regresión para obtener la ecuación de la recta

```python
modelo = LinearRegression() 
modelo.fit(data[['x']], data['y']) 
print ('Ecuación de la recta: y = ', round(modelo.coef_[0],3),'x + ', round(modelo.intercept_,3))#obtenemos la ecuación de la recta
print ('Coeficiente de correlación: ', round(np.corrcoef(data['x'], data['y'])[0,1], 3))#obtenemos el coeficiente de correlación
print ('Coeficiente de determinación: ', round(r2_score(data['y'], modelo.predict(data[['x']])), 3))#obtenemos el coeficiente de determinación
# Gráfica con intervalo de confianza
sns.regplot(x=data['x'], y=data['y'], ci=95, line_kws={"color": "red"}, scatter_kws={"color": "black"})
plt.title('Regresión lineal simple con intervalo de confianza al 95%') 
plt.show()
```
